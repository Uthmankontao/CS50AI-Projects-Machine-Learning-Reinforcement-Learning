Algorithm 1 -- Uniform exploration

```python
my_Bandit = Bandit()
bandits = [[]]
K, T = 10, 500
```

```python
for k in range(K-1):
    bandits.append([])

bandits
```

```python
value = set()
for t in range(T):
    k = np.random.randint(0, K)
    reward = my_Bandit.get_arm(k)
    value.add((k, reward))
    bandits[k].append(reward)
```

```python
fig = plt.figure(figsize=(10,8))
plt.violinplot(bandits)
plt.show()
```
